```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df=pd.read_csv('/content/Iris.csv')
df=df.drop(['Id'],axis=1)
df['Species']=df['Species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}).astype(int)
df.sample(10)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)  # default k value = 5,  but best k value is 3
knn.fit(x_train,y_train)
```

```python
y_train.value_counts()
```

#train performance

```python
train_ans=knn.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_train,train_ans)
```

```python
accuracy_score(y_train,train_ans)
```

```python
print(classification_report(y_train,train_ans))
```

#test performance

```python
y_test.value_counts()
```

```python
test_ans=knn.predict(x_test)
```

```python
confusion_matrix(y_test,test_ans)
```

```python
print(classification_report(y_test,test_ans))
```

```python
accuracy_score(y_test,test_ans)
```

#checking with own data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sl = 1.4
sw = 1.8
pl = 0.04
pw = 0.08

if knn.predict([[sl,sw,pl,pw]])[0]==0:
  print('setosa')
elif knn.predict([[sl,sw,pl,pw]])[0]==1:
  print(f'Versicolor')
else:
  print(f'Verginica')
```

#KNN TEST JSON

```python
import json
```

confusion_matrix() returns a NumPy array â†’ convert to list using .tolist()


classification_report() â†’ use output_dict=True to get dictionary format (instead of string)

accuracy_score() â†’ already a float (JSON supports it)

indent=4 means:

ðŸ‘‰ Add 4 spaces for formatting (pretty printing)
ðŸ‘‰ Make the JSON file more readable
ðŸ‘‰ It does NOT change the data
ðŸ‘‰ It only changes the appearance of the file

Without indent  Your JSON file will look like this (single line)

Why 4?

You can use any number:

indent=2
indent=4
indent=8


2 â†’ 2 spaces

4 â†’ 4 spaces (most commonly used)

```python
# Create performance dictionary
knn_test_performance = {
    "confusion_matrix": confusion_matrix(y_test, test_ans).tolist(),
    "classification_report": classification_report(y_test, test_ans, output_dict=True),
    "accuracy_score": accuracy_score(y_test, test_ans)
}
with open("knn_test_performance.json", "w") as f:
    json.dump(knn_test_performance, f, indent=4)
```
